```python
sample_df = pd.read_csv("SampleSuperstore.csv")
sample_df.drop(columns="Postal Code", inplace=True)

# Select only numeric columns
numeric_df = sample_df.select_dtypes(include=['number'])

# Calculate correlation and covariance
correlation_matrix = numeric_df.corr()
covariance_matrix = numeric_df.cov()

print("Correlation Matrix:")
print(correlation_matrix)

print("\nCovariance Matrix:")
print(covariance_matrix)
```

```python
#top 10 states by most profit
sample_df["State"].unique()
```

```python
state_profit=pd.DataFrame(sample_df.groupby("State")["Profit"].sum())
state_profit.reset_index(inplace=True)


state_profit=state_profit.sort_values(by="Profit",ascending=False)
state_profit.head(10)
```

```python
#most profitable sub category
sub_categories_info=pd.DataFrame(sample_df.groupby("Sub-Category")["Profit"].sum())
sub_categories_info.reset_index(inplace=True)
sub_categories_profit=sub_categories_info.sort_values(by="Profit",ascending=False)
sub_categories_profit.head(10)
```

```python
import matplotlib.pyplot as plt
plt.bar(sub_categories_profit["Sub-Category"],sub_categories_profit["Profit"])
plt.xticks(sub_categories_profit['Sub-Category'] , rotation = "vertical")
plt.xlabel("Sub-Categories")
plt.ylabel("Profit")
plt.show()
```

```python
#Which is most popular sub-category
popular_sub_category=pd.DataFrame(sample_df.groupby("Sub-Category")["Quantity"].sum())
popular_sub_category.reset_index(inplace=True)
popular_sub_category=popular_sub_category.sort_values(by="Quantity",ascending=False)


plt.bar(popular_sub_category["Sub-Category"],popular_sub_category["Quantity"],color="green")
plt.xticks(popular_sub_category["Sub-Category"],rotation="vertical")
plt.title("Sub-Category Vs Quantity")
plt.xlabel("Sub Category")
plt.ylabel("Quantity")
plt.show()
```

```python
#most profitable cutomer segment
customer_segment=sample_df.groupby("Segment").sum()
customer_segment.reset_index(inplace=True)
customer_segment
plt.bar(customer_segment["Segment"],customer_segment["Profit"],color="orange")
plt.xticks(customer_segment["Segment"],rotation="vertical")
plt.title("Segment Vs Profit")
plt.xlabel("Segment")
plt.ylabel("Profit")
plt.show()
```

```python
#highest Sales Volume
city_volume = pd.DataFrame(sample_df.groupby('City')['Sales'].sum())
city_volume =city_volume.sort_values(by="Sales",ascending=False)
city_volume.head()
plt.bar(city_volume.index[0:20], city_volume['Sales'][0:20],color="purple")
plt.xticks(city_volume.index[0:20],rotation = 'vertical')
plt.xlabel("City" )
plt.ylabel("Sales")
plt.show()
```

```python
#sales vs profit based on sub category
sample_df.groupby('Sub-Category')[['Profit', 'Sales']].agg(['sum']).plot.bar()
plt.title("TOTAL PROFIT AND SALES PER SUB CATEGORY")
plt.rcParams['figure.figsize'] = [10, 8]
plt.show()
```

```python
#sales and profit region wise

sample_df.groupby('Region')[['Profit', 'Sales']].agg(['sum']).plot.bar()
plt.title("TOTAL PROFIT AND SALES REGION WISE")
plt.rcParams['figure.figsize'] = [10, 8]
plt.show()
```
